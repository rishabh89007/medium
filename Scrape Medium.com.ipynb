```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import re

# Set up the Chrome driver
options = Options()
options.headless = True  # Run the browser in headless mode
service = Service('C:/Users/risha/Downloads/chromedriver')  # Replace with the path to your chromedriver executable
driver = webdriver.Chrome(service=service, options=options)

# List of URLs to visit
urls = ["https://medium.com/tag/poetry", "https://medium.com/tag/life-lessons", "https://medium.com/tag/technology", 
        "https://medium.com/tag/health", "https://medium.com/tag/writing", "https://medium.com/tag/self-improvement", 
        "https://medium.com/tag/business", "https://medium.com/tag/travel", "https://medium.com/tag/entrepreneurship", 
        "https://medium.com/tag/bitcoin"]

for url in urls:
    # Load the webpage
    driver.get(url)

    # Find the first news title on the page
    news_title = driver.find_element(By.CLASS_NAME, "a.b.c")

    # Extract and print the news title
    text = news_title.text.strip()

    titles = re.findall(r"Member-only\n(.+)", text)

    # Print "Life Lessons" or the tag name based on the URL
    if "poetry" in url:
        print("Poetry")
    elif "life-lessons" in url:
        print("Life Lessons")
    elif "technology" in url:
        print("Technology")
    elif "health" in url:
        print("Health")
    elif "self-improvement" in url:
        print("Self-improvement")
    elif "writing" in url:
        print("Writing")
    elif "business" in url:
        print("Business")
    elif "travel" in url:
        print("Travel")
    elif "entrepreneurship" in url:
        print("Entrepreneurship")
    elif "bitcoin" in url:
        print("Bitcoin")
    print()

    # Print the results
    for title in titles:
        print(title)

    print()

# Quit the driver
driver.quit()
```
